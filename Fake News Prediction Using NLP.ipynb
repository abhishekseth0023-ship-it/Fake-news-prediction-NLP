```python
true_df['label']=0
fake_df['label']=1
```

```python
merged_df = pd.concat([fake_df, true_df], ignore_index=True)
```

```python
merged_df = merged_df.sample(frac=1).reset_index(drop=True)
```

```python
import re
def remove_tags(raw_text):
   new_text=re.sub('<.*>?','',raw_text)
   return new_text
```

```python
merged_df['title']=merged_df['title'].apply(remove_tags)
```

```python
merged_df['text']=merged_df['text'].apply(remove_tags)
```

```python
merged_df
```

```python
merged_df['text']=merged_df['text'].apply(lambda x:x.lower())
```

```python
merged_df['title']=merged_df['title'].apply(lambda x:x.lower())
```

```python
from nltk.corpus import stopwords

sw_text = set(stopwords.words('english'))

merged_df['text'] = merged_df['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in sw_text]))
```

```python
from nltk.corpus import stopwords

sw_text = set(stopwords.words('english'))

merged_df['title'] = merged_df['title'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in sw_text]))
```

```python
x=merged_df.iloc[:,0:2]
y=merged_df['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
```

```python
x_train_tfidf=vec.fit_transform(x_train['text']).toarray()
x_test_tfidf=vec.transform(x_test['text']).toarray()
```

```python
x_train_tfidf=vec.fit_transform(x_train['title']).toarray()
x_test_tfidf=vec.transform(x_test['title']).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model
```

```python
model.fit(x_train_tfidf,y_train)
```

```python
model.score(x_train_tfidf,y_train)
```

```python
model.score(x_test_tfidf,y_test)
```
